```python
df['tags']=df['genre']+df['overview']
```

```python
movies=df[['id','title','tags']]
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd




vectorizer = CountVectorizer(max_features=10000, stop_words='english')
vectors = vectorizer.fit_transform(movies['tags'].values.astype('U')).toarray()


sim = cosine_similarity(vectors)


def recommend(movie_name):

    movie_index = movies[movies['title'] == movie_name].index[0]


    distances = sim[movie_index]


    dist = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])


    print(f"\nMovies similar to '{movie_name}':\n")
    for i in dist[1:6]:
        print(movies.iloc[i[0]].title)


recommend("Dilwale Dulhania Le Jayenge")
```
